```python
import pandas as pd
import numpy as np

school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"
```

```python
student_data = pd.read_csv(student_data_to_load)
school_data = pd.read_csv(school_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

print(school_data_complete.columns.tolist())
```

```python
school_data_complete = pd.read_csv(school_data_to_load)
school_count = school_data_complete["school_name"].count()
total_students = school_data_complete["size"].sum()
total_budget = school_data_complete["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
```

```python
Score_bin = [0, 71, 150]

PassorFail = ['Fail', 'Pass']

student_data['Math Score Percent'] = pd.cut(student_data['math_score'], Score_bin, labels=PassorFail)
student_data['Reading Score Percent'] = pd.cut(student_data['reading_score'], Score_bin, labels=PassorFail)

math_percent = student_data['Math Score Percent'].value_counts(normalize=True).mul(100)
reading_percent = student_data['Reading Score Percent'].value_counts(normalize=True).mul(100)

math_passing = math_percent["Pass"]
reading_passing = reading_percent["Pass"]

Total_Avg = (math_passing + reading_passing) / 2
```

```python
District_Data_table = {'metric':['School Count','Total Students','Total Budget', 'Avg Math Score', 'Avg Reading Score','% Passing Math', '% Passing Reading','Overall Passing Rate'], 
                     'result':[school_count, total_students, total_budget, avg_math_score, avg_reading_score, math_passing, reading_passing, Total_Avg]}
District_Data_table = pd.DataFrame(District_Data_table).round(2) 
District_Data_table
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
grouped_school = school_data_complete.groupby(['school_name'])

total_student = grouped_school.size()

school_type = grouped_school['type'].first()

total_budget = grouped_school['budget'].first()

t_budget_per_student = total_budget/total_student

average_math_score = grouped_school['math_score'].mean()

average_reading_score = grouped_school['reading_score'].mean()

grouped_passing_math = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
percent_passing_math = (grouped_passing_math/total_student)*100

grouped_passing_reading = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
percent_passing_reading = (grouped_passing_reading/total_student)*100

percent_overall_passing = (percent_passing_math + percent_passing_reading)/2


school= pd.DataFrame({
    'School Type': school_type,
    'Total Students':total_student,
    'Total School Budget': total_budget,
    'Per Student Budget': t_budget_per_student,
    'Average Math Score': average_math_score,
    'Average Reading Score': average_reading_score,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    '% Overall Passing Rate': percent_overall_passing})

school = pd.DataFrame(school).round(2) 

school
```

```python
top_schools = school.sort_values(by='% Overall Passing Rate',ascending=False)
top_schools.head()
```

```python
bottom_schools = school.sort_values(by='% Overall Passing Rate')
bottom_schools.head()
```

```python
Avg_Math_9th = school_data_complete[school_data_complete['grade']=='9th'].groupby('school_name')['math_score'].mean()
Avg_Math_10th = school_data_complete[school_data_complete['grade']=='10th'].groupby('school_name')['math_score'].mean()
Avg_Math_11th = school_data_complete[school_data_complete['grade']=='11th'].groupby('school_name')['math_score'].mean()
Avg_Math_12th = school_data_complete[school_data_complete['grade']=='12th'].groupby('school_name')['math_score'].mean()

math_scores_by_grade = pd.DataFrame({
    '9th Grade': Avg_Math_9th,
    '10th Grade':Avg_Math_10th,
    '11th Grade': Avg_Math_11th,
    '12th Grade': Avg_Math_12th})

math_scores_by_grade
```

```python
Avg_Reading_9th = school_data_complete[school_data_complete['grade']=='9th'].groupby('school_name')['reading_score'].mean()
Avg_Reading_10th = school_data_complete[school_data_complete['grade']=='10th'].groupby('school_name')['reading_score'].mean()
Avg_Reading_11th = school_data_complete[school_data_complete['grade']=='11th'].groupby('school_name')['reading_score'].mean()
Avg_Reading_12th = school_data_complete[school_data_complete['grade']=='12th'].groupby('school_name')['reading_score'].mean()

reading_scores_by_grade = pd.DataFrame({
    '9th Grade': Avg_Reading_9th,
    '10th Grade':Avg_Reading_10th,
    '11th Grade': Avg_Reading_11th,
    '12th Grade': Avg_Reading_12th})

reading_scores_by_grade
```

```python
spending_bins = [0, 585, 615, 645, 675]
spending_ranges = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school["Spending Ranges Per Student"] = pd.cut(t_budget_per_student, spending_bins, labels = spending_ranges)
```

```python
spending_math_score = school.groupby(["Spending Ranges Per Student"]).mean()['Average Math Score']
spending_reading_score = school.groupby(["Spending Ranges Per Student"]).mean()['Average Reading Score']
spending_percent_passing_math = school.groupby(["Spending Ranges Per Student"]).mean()['% Passing Math']
spending_percent_passing_reading = school.groupby(["Spending Ranges Per Student"]).mean()['% Passing Reading']
spending_overall_passing = school.groupby(["Spending Ranges Per Student"]).mean()['% Overall Passing Rate']

scores_by_school_spend = pd.DataFrame
    'Math Score by Spending': spending_math_score,
    'Reading Score by Spending':spending_reading_score,
    'Math Pass % by Spending': spending_percent_passing_math,
    'Reading Pass % by Spending': spending_percent_passing_reading,
    'Overall Pass % by Spending': spending_overall_passing})

scores_by_school_spend
```

```python
size_bins = [0, 1000, 2000, 5000]
size_ranges = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school["Total Students"] = pd.cut(total_student, size_bins, labels = size_ranges)
```

```python
size_math_score = school.groupby(["Total Students"]).mean()['Average Math Score']
size_reading_score = school.groupby(["Total Students"]).mean()['Average Reading Score']
size_percent_passing_math = school.groupby(["Total Students"]).mean()['% Passing Math']
size_percent_passing_reading = school.groupby(["Total Students"]).mean()['% Passing Reading']
size_overall_passing = school.groupby(["Total Students"]).mean()['% Overall Passing Rate']

scores_by_school_size = pd.DataFrame({
    'Math Score by size': size_math_score,
    'Reading Score by size':size_reading_score,
    'Math Pass % by size': size_percent_passing_math,
    'Reading Pass % by size': size_percent_passing_reading,
    'Overall Pass % by size': size_overall_passing})

scores_by_school_size
```
